# Classification

```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
X,y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```
